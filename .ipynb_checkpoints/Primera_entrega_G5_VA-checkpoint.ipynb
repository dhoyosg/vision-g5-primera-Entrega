## Punto 1 literal a

```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
##Generamos una matriz de ceros de 400*400
M = np.zeros((400,400,3),np.uint8)
##Manipulamos los valores de los canales para obtener el resultado esperado

#Amarillo azul en 0
M[0:,:100,0] = 255
M[0:,:100,1] = 255

##cyan=rojo en 0
M[0:,100:200,1] = 255
M[0:,100:200,2] = 255

##rojo= azuly verde en 0
M[0:,200:300,0] = 255

##Verde rojo y azul en cero.
M[0:,300:400,1] = 255
##M ya está en el canal rgb, por lo que no toca hacer el cambio de canales. Adicionalmente ya está creada,
#por lo que no es necesario reabrirla con cv2.
plt.imshow(M)
```

## Punto 1 literal b

```python
##Tenemos ya a M como rgb
#Descomposición en varios canales del espacio de color RGB y sus respectivos histogramas
img_R = M[:,:,0]
img_G = M[:,:,1]
img_B = M[:,:,2]

##para sacar el lab, hacemos el cambio de rgb a lab, usando cv2.
MenLAB=cv2.cvtColor(M,cv2.COLOR_RGB2LAB)
##extraemos cada canal LAB
img_L = MenLAB[:,:,0]
img_A =  MenLAB[:,:,1]
img_BB = MenLAB[:,:,2]

fig, arreglo_plots = plt.subplots(6,2, figsize=(25, 25))
colors = ('r','g','b')
#Canal RGB
arreglo_plots[0,0].set_title('Canal R')
arreglo_plots[0,0].imshow(img_R, cmap="gray")
arreglo_plots[0,1].set_title('Histograma Canal R')
M_array_i = img_R.ravel()
arreglo_plots[0,1].hist(M_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[0])

arreglo_plots[1,0].set_title('Canal G')
arreglo_plots[1,0].imshow(img_G, cmap="gray")
arreglo_plots[1,1].set_title('Histograma Canal G')
M_array_i = img_G.ravel()
arreglo_plots[1,1].hist(M_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[1])
    
arreglo_plots[2,0].set_title('Canal B')
arreglo_plots[2,0].imshow(img_B, cmap="gray")
arreglo_plots[2,1].set_title('Histograma Canal B')
M_array_i = img_B.ravel()
arreglo_plots[2,1].hist(M_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[2]);

#Canal LAB
arreglo_plots[3,0].set_title('Canal L')
arreglo_plots[3,0].imshow(img_L, cmap="gray", vmin = 0, vmax = 255)
arreglo_plots[3,1].set_title('Histograma Canal L')
M_array_i = img_L.ravel()
arreglo_plots[3,1].hist(M_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[0])

arreglo_plots[4,0].set_title('Canal A')
arreglo_plots[4,0].imshow(img_A, cmap="gray", vmin = 0, vmax = 255)
arreglo_plots[4,1].set_title('Histograma Canal A')
M_array_i = img_A.ravel()
arreglo_plots[4,1].hist(M_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[1])
    
arreglo_plots[5,0].set_title('Canal B')
arreglo_plots[5,0].imshow(img_BB, cmap="gray", vmin = 0, vmax = 255)
arreglo_plots[5,1].set_title('Histograma Canal B')
M_array_i = img_BB.ravel()
arreglo_plots[5,1].hist(M_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[2]);
```

### Conclusiones: 


#### Espacio RGB
Los valores presentes en los canales R, G y B corresponden a los valores asignados en el literal a. 

-Canal R: Se observa que la franja de 0-100 y la de 200-300 pixeles se encuentran saturadas en 255 para este canal, mientras que las demás se encuentran en 0. Esto corresponde al histograma, donde se ve que la mitad de los pixeles tienen un nivel de saturación de 0 y la otra mitad tiene un nivel de saturación de 255.

-Canal G: La franja vertical de 200-300 pixeles es la única porción de la imagen que se ve negra. Esto se debe a que el color de esta porción de la imagen (rojo) fue el único que se obtuvo manteniendo la saturación del canal verde en 0, mientras que para las otras porciones de la imagen la saturación del canal verde fue de 255. Lo cual corresponde al histograma, donde el 75% de los pixeles tienen una saturación de 255 y el 25% restante tiene una saturación de 0.

-Canal B: El color cian, correspondiente a la franja vertical entre 100 y 200 pixeles fue la única porción de la imagen en la que la saturación del canal B era 255 en vez de 0. 
Es por esto por lo que en el histograma el 25% de los pixeles tienen una saturación de 255 y el 75% restante tiene una saturación de 0.

El espacio de color LAB está basado en cuánto color se encuentra en el eje rojo-verde (a), en el eje azul-amarillo (b), y su iluminación (L).

Se usó la calculadora de color disponible en la página http://colorizer.org/ para encontrar los siguientes valores y poder comprender mejor los histogramas.

| Color | L   | A | B   | 
|------|------|------|------|
|Amarillo| 247|106|222|
|  Cian  | 230|80|114|
|  Rojo  | 135|208|195|
|Verde| 221|42|211|



-Canal L: Contiene la información de la iluminación de la foto. Se podría ver como la imagen en escala de grises. 

En el histograma se puede observar cómo un 25% de los pixeles se encuentran aproximadamente en cada valor de iluminación de la columna L de la tabla.

-Canal A: Contiene la información de cuanto color se encuentra en el eje verde-rojo (con 0=verde y 255=rojo). 
Se esperaría que la franja roja de la imagen fuera blanca (255) en el canal A y que la franja verde tenga una intensidad de 0 y se visualizara negra, lo cual no ocurre. Al revisar la tabla presentada anteriormente se encuentra que estos tonos específicos corresponderían a 208 y 42 aproximadamente, y usando la calculadora de color se puede observar que son diferentes a los tonos de rojo y verde para los cuales el canal A es 0 y 255.
En el histograma se puede observar cómo un 25% de los pixeles se encuentran aproximadamente en cada valor de la columna A de la tabla.

-Canal B: Contiene la información de cuanto color se encuentra en el eje azul-amarillo (con 0=azul y 255=amarillo). 
Al igual que con el canal A, se esperaría que la franja amarilla tuviera valores blancos, pero de la tabla y de la calculadora de color se tiene que esta tonalidad no es igual a la de máxima intensidad en el canal B, pero tiene un valor bastante cercano (247).

## Punto 1 literal c.

Lea la imagen "eiffel.jpg" que se encuentra en la carpeta "res", redimensionela a 800x600 píxeles, muestre la imagen y su histograma asociado.



```python
eif = cv2.imread("res/eiffel.jpg", cv2.IMREAD_COLOR)
##redimensionamos a 800*600, recordemos que está en BGR
torre=cv2.resize(eif,(800, 600), interpolation = cv2.INTER_NEAREST)

##la pasamos de BGR a HSV
torre_HSV  = cv2.cvtColor(torre, cv2.COLOR_BGR2HSV)
torre_H=torre_HSV[:,:,0]
torre_S=torre_HSV[:,:,1]
torre_V=torre_HSV[:,:,2]
##La obtenemos en RGB para mostrarla
torre_RGB=cv2.cvtColor(torre, cv2.COLOR_BGR2RGB)


##Mostrar el histograma

fig, arreglo_plots = plt.subplots(2,1, figsize=(20, 10))


#Canal RGB
arreglo_plots[0].set_title('Imagen RGB')
arreglo_plots[0].imshow(torre_RGB)
arreglo_plots[1].set_title('Histograma RGB')
color = ('r','g','b')
for i,col in enumerate(color):
    histr = cv2.calcHist([torre_RGB],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()

##Mostrar el HSV

fig, arreglo_plots = plt.subplots(3,2, figsize=(25, 25))
colors = ('r','g','b')
#Canal RGB
arreglo_plots[0,0].set_title('Canal H')
arreglo_plots[0,0].imshow(torre_H, cmap="gray")
arreglo_plots[0,1].set_title('Histograma Canal H')
M_array_i = torre_H.ravel()
arreglo_plots[0,1].hist(M_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[0])

arreglo_plots[1,0].set_title('Canal S')
arreglo_plots[1,0].imshow(torre_S, cmap="gray")
arreglo_plots[1,1].set_title('Histograma Canal S')
M_array_i = torre_S.ravel()
arreglo_plots[1,1].hist(M_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[1])
    
arreglo_plots[2,0].set_title('Canal V')
arreglo_plots[2,0].imshow(torre_V, cmap="gray")
arreglo_plots[2,1].set_title('Histograma Canal V')
M_array_i = torre_V.ravel()
arreglo_plots[2,1].hist(M_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[2]);
```

### Conclusiones

Estudiando el espacio de color matiz-saturación-valor (HSV), se puede concluir:

El canal H es una representación radial del canal RGB, así que escalando tenemos; 0 pixeles representan al rojo, 85 representa al verde, 240 al azul y 255 de nuevo al rojo, donde los valores intermedios significan combinaciones del espacio de color RGB. Para esta imagen se puede observar que hay predominancia en valores cercanos a 100 que representa valores relacionados al verde para los arboles, valores que representan al cian, cercanos al 120 que se observa en el cielo y en la fuente, y para la torre hay predominancia de pixeles en el rojo.

El canal S indica que tan saturada está la imagen, donde 0 significa que la imagen está completamente saturado, es decir la imagen es blanca o 255 que el color es puro en cuanto a colorimetría, según el histograma se puede observar que la imagen se encuentra en su mayor parte saturada, es decir la pureza del color es baja para los tonos azules del cielo, las nubes y la fuente, para los verdes la saturación es la más alta, y para la torre el color se ecuentra en su mayor pureza.

El canal V indica el nivel de intensidad de una imagen, donde 0 significa que la intensidad es la más baja, es decir una imagen en negro y 255 que el color es completamente puro. Según el histograma la mayoría de los pixeles pertenecen a colores con altas intensidades, como el cielo, la torre, la fuente y el parque, y los pixeles que corresponden a las más bajas intensidades son los que pertenecen al bosque.

```python
##Literal A punto 1
#A continuación, se presenta una transformación sobre ambas imágenes
##La transformación es: (1-k)*I1 + KI2
##Multiplicar ambas imagenes por las constantes necesarias y posteriomente sumarlas
##tienen igual tamaño?
#print(colombia.shape)
#print(japon.shape)
##Como no poseen igual tamaño, entonces le hago resize a la segunda, para poderlas sumar en la transormación. CV2 provee función
##Shape me da las dimensiones de colombia, pos 0 las filas, pos 1 columnas.
japon = cv2.resize(japon, (colombia.shape[1], colombia.shape[0]))
print("Imagenes ahora tienen el mismo tamaño")
print("Colombia: ",colombia.shape)
print("Japon: ",japon.shape)

##creo función para ahorrar código: Recibe el valor de K y ambas imagenes, realiza la transformación y retorna el resultado
def transformacion(k,colombia,japon):
    #creo auxiliares para procesar canales
    colombiaaux=colombia
    #separo los canales para multiplicarles
    colombiaaux[:,:,0]=cv2.multiply((1-k),(colombiaaux[:,:,0]))
    colombiaaux[:,:,1]=cv2.multiply((1-k),(colombiaaux[:,:,1]))
    colombiaaux[:,:,2]=cv2.multiply((1-k),(colombiaaux[:,:,2]))
    japonaux=japon
    #separo los canales para multiplicarles
    japonaux[:,:,0]=cv2.multiply((k),(japonaux[:,:,0]))
    japonaux[:,:,1]=cv2.multiply((k),(japonaux[:,:,1]))
    japonaux[:,:,2]=cv2.multiply((k),(japonaux[:,:,2]))
    resultado=cv2.add(colombiaaux,japonaux)
    return resultado

#con k=0,5
resultado1=transformacion(0.5,colombia.copy(),japon.copy())

#con k=1
resultado2=transformacion(1,colombia.copy(),japon.copy())

#con k=0
resultado3=transformacion(0,colombia.copy(),japon.copy())

##plot de resultados, arreglo de 3*3
fig,((x1,x2,x3),(y1,y2,y3),(z1,z2,z3))=plt.subplots(3,3, figsize=(30,20))
fig.suptitle("Transformaciones resultantes",fontsize=30)
x1.set_title("Colombia",fontsize=30)
x2.set_title("Japón",fontsize=30)
x3.set_title("Con k=0.5",fontsize=20)
x1.imshow(colombia, cmap='gray', aspect='auto')
x2.imshow(japon, cmap='gray', aspect='auto')
x3.imshow(resultado1, cmap='gray', aspect='auto')

y3.set_title("Con k=1",fontsize=20)
y1.imshow(colombia, cmap='gray', aspect='auto')
y2.imshow(japon, cmap='gray', aspect='auto')
y3.imshow(resultado2, cmap='gray', aspect='auto')

z3.set_title("Con k=0",fontsize=20)
z1.imshow(colombia, cmap='gray', aspect='auto')
z2.imshow(japon, cmap='gray', aspect='auto')
z3.imshow(resultado3, cmap='gray', aspect='auto')
```
