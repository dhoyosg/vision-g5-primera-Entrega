## 2.

Lea la imagen "gato.jpg" que se encuentra en la carpeta "res" en RGB. A continuación, se muestran dos funciones para creación e implementación de kernels.

```python
#Función para hacer el kernel, su parámetro es el tamaño de éste
def make_kernel(size):
    #Matriz de unos del tamaño de entrada, datos de tipo flotante
    kernel = np.ones((size,size),dtype=np.float64)
    #r corresponde al entero más grande que sea menor o igual a size/2
    r = math.floor(size/2)
    return (r,kernel/(size**2))

#Función de aplicación de kernel sobre la imagen bajo los parámetros 
#imagen de entrada, máscara, valores en x,y,kernel,r
def apply_kernel(img, res, x,y,kernel,r):
    #Recorrer la imagen en el intervalo del radio del kernel
    for i in range(-r, r+1):
        for j in range(-r, r+1):
            #Convolución sobre cada pixel de la imagen
            res[x][y] += img[x+i][y+j] * kernel[r+i][r+j]
```

La función make_kernel sirve para crear a la matriz de convolución o también llamada máscara de convolución, basicamente sirve para definir el tamaño de la matriz que se va a aplicar a la imagen, es decir el radio del kernel.
La función apply_kernel tiene como función después de definir el kernel, aplicar a cada pixel la matriz de convolución. Esta función hace una sumatoria de la multiplicación de los pixeles del vecindario del pixel de la imagen, con los valores del kernel, asignandole un nuevo valor al pixel que se quiere modificar, y esto mismo se le hace a todos los pixeles de toda la imagen.
En la convolución más sencilla los bordes de la imagen quedan con valores de 0, esto sucede porque como no están definidos los valores después del borde, estos pixeles quedan asignados con ningun valor, es decir con el cero.

## 3.

Lea la imagen "gato.jpg" que se encuentra en la carpeta "res" en RGB. En este ejercicio usted debe recrear distintos tipos de ruido que se presentan en imágenes.

El ruido es un factor que se encuentra en la mayoría de los problemas de visión artificial, utilizando las funciones de ruido Gaussiano, Pimienta y Poisson, obtenga:

<img src = "res/gatoxf.png">

A continuación se muestra la implementación de las funciones:

```python
#Definir la función ruido gaussiano
def gaussian_noise(img, std=0.02):
    #Las variables row, col y ch dependen de la imagen de entrada y sus componentes
    mean = 0
    row = img.shape[0]
    col = img.shape[1]
    ch =  img.shape[2] if len(img.shape) == 3 else 1       
    #Definir la matriz de ceros noisy  del mismo tamaño que la imagen original
    noisy = np.zeros(img.shape, np.float64)
    #Generar una distribución normal a partir del comando de numpy
    gauss = np.random.normal(mean,std,(row,col,ch)) if ch > 1 else np.random.normal(mean,std,(row,col))
    gauss = gauss.reshape(row,col,ch)               if ch > 1 else gauss.reshape(row,col)
    #La imagen con ruido gaussiano es la suma de la imagen de entrada con la procesada
    noisy =  img + gauss
    #Ajustar los datos para que estén entre 0 y 255
    noisy[noisy<0] = 0
    noisy[noisy>255] = 255
    return noisy

#Definir la función de ruido impulsivo
def sp_noise(img, s_vs_p, amount):
    
    #Obtener los valores de row, col y ch con base en la imagen de entrada
    row = img.shape[0]
    col = img.shape[1]
    ch =  img.shape[2] if len(img.shape) == 3 else 1  
    
    #Crear una copia de la imagen para volverla ruidosa
    noisy = img.copy()
    
    # Salt mode
    
    #Multiplica los parámetros con el tamaño de la imagen, los redondea al mínimo entero
    num_salt = np.ceil(amount * img.size * s_vs_p)
    #Genera el vector coordenadas en la imagen de copia usando la función randint de numpy
    coords = [np.random.randint(0, i - 1, int(num_salt))
          for i in img.shape]

    #La imagen evaluada en el vector coords se satura
    noisy[coords] = 255
    
    # Pepper mode
    
    #Se multiplican los parámetros por el tamaño de la imagen, variando s_vs_p
    num_pepper = np.ceil(amount* img.size * (1. - s_vs_p))
    #Generar el vector coords utilizando randint
    coords = [np.random.randint(0, i - 1, int(num_pepper))
          for i in img.shape]
    #La imagen en el vector coords se satura en color negro
    noisy[coords] = 0
    return noisy

#Definir la función ruido de Poisson
def poisson_noise(img):
    
    #Se aplica el comando de numpy random.poisson a la imagen de entrada
    noisy = np.random.poisson(img) 
    #Los valores deben estar entre 0 y 255
    noisy[noisy<0] = 0
    noisy[noisy>255] = 255
    return noisy
```

## Sugerencia: Ajuste los parámetros std = 50, s_vs_p =  0.5, amount = 0.05

## No olvide variar los parámetros y observar qué ocurre al hacerlo.

```python

img = cv2.cvtColor(cv2.imread('res/gato.jpg'), cv2.COLOR_BGR2RGB)
imgg = gaussian_noise(img, 70)
imgsp = sp_noise(img, 0.5, 0.05)
imgp = poisson_noise(img)

fig, arreglo_plots = plt.subplots(2,2, figsize=(15, 15))

arreglo_plots[0][0].imshow(img,cmap='gray')
arreglo_plots[0][0].set_title('Imagen original')

arreglo_plots[0][1].imshow(imgg,cmap='gray')
arreglo_plots[0][1].set_title('Ruido gausiano')

arreglo_plots[1][0].imshow(imgsp,cmap='gray')
arreglo_plots[1][0].set_title('Ruido SP')

arreglo_plots[1][1].imshow(imgp,cmap='gray')
arreglo_plots[1][1].set_title('Ruido Poisson')
```

## Conclusiones:

El ruido en una imagen tiende a ser información no desada, y existen varios tipos de ruidos, que son los que se observan en las imagenes anteriores:

La primer imagen tiene ruido Gaussiano con los parámetros std = 50, s_vs_p =  0.5, amount = 0.05; según la función Gaussiana, los pixeles de la imagen original cambian sus valores de forma aleatoría con tendencia a una mediana de distribución, asociado a la campana Gaussiana, por eso si disminuye el std se obtienen valores que tienden mas al blanco.

La segunda imagen tiene un ruido llamado sal y pimienta el cual modifica pocos valores de la imagen total, los cuales no coinciden con valores de la imagen original. Por eso se nota mas el tamaño del ruido, pero no se distorsiona tanto la imagen como en el ruido Gaussiano que es aleatorio y cambia más pixeles.

La última imagen, tiene ruido de Poisson o ruido de disparo, es una función también aleatoria, similiar en ciertas condiciones a la distribución Gaussiana, la cuál modifica los valores de los pixeles, con valores cercanos en intensidad, por lo cual no se miran mayores cambios en el ruido.

Cuando aplique 100 veces el filtro de media 3x3 debe obtener lo siguiente:

###  Implementación del filtro:

```python
####Implementación 100 veces filtro####
#tamaño del kernel
k_size = 3
#crear el kernel
kernel = np.ones((k_size,k_size),np.float64)/(k_size**2)
img = cv2.imread('res/rice.png')
#crear imagen que se va a modificar las n veces que se aplique el filtro
res2 = img.copy()
#ciclo que implementa el filtro de media
for i in range(0,100):
    res2 = cv2.filter2D(res2, ddepth=-1, kernel= kernel, borderType=cv2.BORDER_REPLICATE)

fig = plt.figure(figsize=(20,24))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title("Imagen Original")

plt.subplot(1,2,2)
plt.imshow(res2,cmap='gray')
plt.title("Imagen Salida")
```

```python
####Implementación 100 veces filtro####
#tamaño del kernel
k_size = 11
#crear el kernel
kernel = np.ones((k_size,k_size),np.float64)/(k_size**2)
img = cv2.imread('res/rice.png')
#crear imagen que se va a modificar las n veces que se aplique el filtro
res2 = img.copy()
#ciclo que implementa el filtro de media
for i in range(0,100):
    res2 = cv2.filter2D(res2, ddepth=-1, kernel= kernel, borderType=cv2.BORDER_REPLICATE)

fig = plt.figure(figsize=(20,24))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title("Imagen Original")

plt.subplot(1,2,2)
plt.imshow(res2,cmap='gray')
plt.title("Imagen Salida")
```

## Conclusión: 

El filtro lineal de media aplica un promedio de los valores de la vecindad, al valor de pixel que se desea cambiar. Al aplicar varias veces el mismo kernel a una imagen, empieza a perderse las fronteras de color, por ejemplo para esta imagen si el vecino del blanco es el negro, en promedio será gris, y así al ir aplicando más filtro de media, esta frontera cromática va aumentando, difimunando más la imagen y perdiendo definición. Al aumentar el tamaño del kernel, las fronteras en la imagen se van perdiendo más rápido, porque aumentan los pixeles del vecindario que se le aplica. Por eso la segunda imagen se mira más borrosa, a pesar de aplicar la misma cantidad de veces el filtro.

```python
####Implementación 100 veces filtro####
#tamaño del kernel
k_size = 3
#crear el kernel
kernel = np.ones((k_size,k_size),np.float64)/(k_size**2)
img = cv2.imread('res/rice.png')
#crear imagen que se va a modificar las n veces que se aplique el filtro
res2 = img.copy()
#ciclo que implementa el filtro de media
for i in range(0,2):
    res2 = cv2.filter2D(res2, ddepth=-1, kernel= kernel, borderType=cv2.BORDER_REPLICATE)

fig = plt.figure(figsize=(20,24))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title("Imagen Original")

plt.subplot(1,2,2)
plt.imshow(res2,cmap='gray')
plt.title("Imagen Salida")
```

## Conclusión: 

Ahora sólo se aplicaron 2 veces el filtro de media a la imagen original, por lo cuál los valores en las fronteras de los granos se ven un poco difuminados; si se aumenta la cantidad de filtros, aumenta la frontera de grises, porque van aumentando los pixeles promediados; y si se aumentan infinitos filtros, el resultado final, será que toda la imagen es correspondientes al promedio de todos sus pixeles, osea será una imagen del mismo color.
